```python
df_numeric = df_num_clm[df_num_clm >=0]
df_numeric
```

```python
for i in df_numeric:
    df_numeric[i] = df_numeric[i].replace(np.nan , df_numeric[i].median())
# df_numeric['Income'] = df_numeric['Income'].replace(np.nan , df_numeric['Income'].median())
```

```python
df_cat = df.select_dtypes(include = ["object"])
```

```python
for i in df_cat:
    print(df_cat[i].value_counts())
    print("------------------------------------------------------------------------")
```

```python
label_encoding = ["Education_Level"]
freq_encoder = ["Housing_Status"]
freq_encoder
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in label_encoding:
    le = LabelEncoder()
    arr = le.fit_transform(df_cat[i])
    df_cat[i] = pd.DataFrame(arr, columns=[i])
```

```python
for i in freq_encoder:
    s = df_cat[i].value_counts()
    df_cat[i] = df_cat[i].map(s)
```

```python
df_concat = pd.concat([df_numeric , df_cat], axis = 1)
```

```python
df_concat
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ss = MinMaxScaler()
```

```python
ss.fit(df_concat)
```

```python
df_concat.describe().round(2)
```

```python
arr = ss.transform(df_concat).round(2)
arr
```

```python
for i in df_concat:
    arr2 = pd.DataFrame(arr, columns =[df_concat.columns]).round(2)
arr2
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X = arr2.drop(columns = ["Default"])
```

```python
y = arr2["Default"]
```

```python
print(y.value_counts())
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.5 ,stratify=y, random_state=12)
```

```python
y_train = y_train.astype(int)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train, y_train.values.ravel())
```

```python
lr.score(X_test, y_test)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, classification_report
```

```python
y_pred=lr.predict(X_test)
```

```python
print('P_score', precision_score(y_test, y_pred))
```

```python
macro_precision = precision_score(y_test, y_pred, average='macro', zero_division=0)
```

```python
macro_precision
```

```python
print('R_score', recall_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
